```python
#!/usr/bin/env python
# coding: utf-8

# In[1]:

import urllib.request, json

import requests
import pandas as pd
import io
import json
import os
import zipfile

# In[2]:

#define function for API to download statcan tables
def StatcanTable(TableNumber):
    TableNumber = str(TableNumber)
    r = requests.get("https://www150.statcan.gc.ca/n1/en/tbl/csv/" + TableNumber + "-eng.zip", stream=True)
    print(r)
    z = zipfile.ZipFile(io.BytesIO(r.content))
    z.extractall(os.getcwd())
    RawData = pd.read_csv(TableNumber + ".csv",low_memory=False)
    return RawData
```

```python
# Indigenous CSD Census 2021

RawData_ind = StatcanTable(98100266)
Indigenous21 = RawData_ind[['REF_DATE','GEO','DGUID','Statistics (3)','Indigenous identity (9):Total - Indigenous identity[1]']]
Indigenous21.rename(columns={'REF_DATE': 'Year', 'GEO':'Geography', 'DGUID': 'ID', 'Statistics (3)':'Statistics','Indigenous identity (9):Total - Indigenous identity[1]':'Count'}, inplace = True)
Indigenous21
```

```python
#Visible Minority CMA Census 2021

RawData_vis = StatcanTable(98100308)
VisibleMinority21 = RawData_vis[['REF_DATE','GEO','DGUID','Statistics (3)','Visible minority (15)','Immigrant status and period of immigration (11):Total - Immigrant status and period of immigration[1]']]
VisibleMinority21.rename(columns={'REF_DATE': 'Year', 'GEO':'Geography', 'DGUID': 'ID', 'Statistics (3)':'Statistics','Visible minority (15)':'Visible Minority','Immigrant status and period of immigration (11):Total - Immigrant status and period of immigration[1]':'Count'}, inplace = True)
VisibleMinority21
```
